```python
Zip = pd.read_pickle("./datasets/zip_demo.p")
Zip
```

```python
fully_integrated_dataframe = census.merge(land_use , on = 'ward')\
                                   .merge(Zip , on = 'zip')
fully_integrated_dataframe.sort_values(by = 'ward' , ascending = False)
```

```python
fully_integrated_dataframe.merge(license , on = 'zip' , how = 'inner' , suffixes=['_X' , "_Y"])
```
